# 데이터 확인

```python
import pandas as pd

# 파일 패스 설정
file_paths = {
    "금융 접근성": "/Users/jochaeyeon/Desktop/4-2/캡스톤_행정/capstondesign_-administration/금융 접근성.xlsx",
    "금융 공기업": "/Users/jochaeyeon/Desktop/4-2/캡스톤_행정/capstondesign_-administration/금융공기업.xlsx",
    "금융 취약계층": "/Users/jochaeyeon/Desktop/4-2/캡스톤_행정/capstondesign_-administration/금융취약계층.xlsx",
}

# 데이터 불러오기
dfs = {key: pd.read_excel(path) for key, path in file_paths.items()}

# 데이터 확인
for key, df in dfs.items():
    print(f"✅ {key} 데이터 확인:")
    print(df.head(), "\n")
```

# 데이터 전처리

```python
# 1) 날짜 형식 변환 (YYYYMMDD → YYYY-MM-DD)
def convert_date_format(df):
    df['일자'] = pd.to_datetime(df['일자'], format='%Y%m%d', errors='coerce')
    return df

#  2) 중복 기사 제거 (제목 & URL 기준)
def remove_duplicates(df):
    df = df.drop_duplicates(subset=['제목', 'URL'], keep='first')
    return df

# 3) 불필요한 기사 필터링 (광고, 인사 기사 삭제)
def filter_unwanted_articles(df):
    unwanted_keywords = ["인사", "공지", "광고", "채용", "발표"]
    df = df[~df['제목'].str.contains('|'.join(unwanted_keywords), na=False)]
    return df

# 4) 결측값 처리 (NaN → 빈 문자열)
def handle_missing_values(df):
    df = df.fillna("")
    return df

#  전처리 실행 (각 데이터셋별로 진행)
for key in dfs:
    dfs[key] = convert_date_format(dfs[key])  # 날짜 변환
    dfs[key] = remove_duplicates(dfs[key])    # 중복 기사 제거
    dfs[key] = filter_unwanted_articles(dfs[key])  # 불필요한 기사 제거
    dfs[key] = handle_missing_values(dfs[key])  # 결측값 처리

# 정리된 파일 저장
for key, df in dfs.items():
    cleaned_filename = f"/Users/jochaeyeon/Desktop/4-2/캡스톤_행정/capstondesign_-administration/{key}_정리된.xlsx"
    df.to_excel(cleaned_filename, index=False)
    print(f"저장 완료: {cleaned_filename}")
```
