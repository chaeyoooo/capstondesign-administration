```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc  

df = pd.read_csv("/Users/jochaeyeon/Desktop/4-2/_/capstondesign_-administration/.csv")


rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

type_labels = {
    1: '',
    2: '',
    3: '',
    4: '',
    5: '',
    0: ''
}
df[''] = df[''].map(type_labels)


pivot = df.pivot_table(index='', columns='', values='', aggfunc='count', fill_value=0)


pivot.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('     ', fontsize=18)
plt.xlabel('')
plt.ylabel('')
plt.legend(title='')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
