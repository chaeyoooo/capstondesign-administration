```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rc  

df = pd.read_csv("/Users/jochaeyeon/Desktop/4-2/캡스톤_행정/capstondesign_-administration/최종.csv")


rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False  

type_labels = {
    1: '대출·자금지원',
    2: '금융교육',
    3: '채무조정·신용회복',
    4: '기부·장학·생활지원',
    5: '디지털',
    0: '기타'
}
df['유형명'] = df['유형코드'].map(type_labels)


pivot = df.pivot_table(index='년도', columns='유형명', values='정책', aggfunc='count', fill_value=0)


pivot.plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title('연도별 금융취약계층 지원 정책 유형 분포', fontsize=18)
plt.xlabel('년도')
plt.ylabel('건수')
plt.legend(title='지원유형')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
